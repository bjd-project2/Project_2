```python
import pandas as pd
import numpy as np
#MySql stuff
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()


#CSVs
parks_csv = "parks.csv"
species_csv = "species.csv"
```

```python
park_df = pd.read_csv(parks_csv)
species_df = pd.read_csv(species_csv, low_memory =False)
```

```python
park_df = park_df[['Park Name' , 'State' , 'Latitude' , 'Longitude'
]]
park_df.head()
```

```python
species_df = species_df[['Park Name' , 'Common Names' , 'Scientific Name' , 
                         'Category' , 'Abundance' , 'Nativeness' , 'Conservation Status']]
species_df.head(3745)
```

```python
#Clean Common names removed NaN names
filtered_df = species_df.replace(to_replace='None', value=np.nan)
name_filtered_df = filtered_df[filtered_df['Common Names'].notnull()]

#Clean Con
```

```python
#SQL engine
engine = create_engine("mysql://root:April131994@localhost/parks_species_db")
```

```python

park_df.to_sql('parksdf', engine, schema= 'parks_species_db', if_exists='replace', index=True)

species_df.to_sql('speciesdf', engine, schema= 'parks_species_db', if_exists='replace', index=True)
```

```python
#Group By For Bora

# Using GroupBy in order to separate the data into fields according to "state" values
grouped_speciesDF = species_df.loc[species_df['Abundance'] == "Rare", :]

# The object returned is a "GroupBy" object and cannot be viewed normally...
print(grouped_speciesDF)

# In order to be visualized, a data function must be used...
grouped_speciesDF.head(10)
```
